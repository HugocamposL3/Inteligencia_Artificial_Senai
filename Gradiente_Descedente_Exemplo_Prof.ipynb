```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def cal_custo(w,x,y):
  m=len(y)
  desejado=np.dot(x,w) ## "np.dot" é uma multiplicação entre vetores
  custo= ((1/2*m)*np.sum(np.square(desejado-y)))
  return custo
```

```python
x = 2*np.random.rand(100,1)
y = 5+2*x+np.random.randn(100,1)
```

```python
plt.plot(x)
plt.plot(y)
```

```python
plt.scatter(x,y)
```

```python
def gradiente (x,y,w,learning_rate = 0.01, iteration = 100): # x = entrada , y = saida , learning rate "taxa de aprendizagem" (alfa), interação = quantas vezes ele vai fazer o gradiente
  m = len(y)
  custo_historico = np.zeros(iteration) # criar um vetor cheio de zeros
 

  for i in range(iteration):
    #multiplicação matricial resultando em uma única saída de tamanho 2 por n (2:n) m por 2 (m:2)
    saida = np.dot(x,w)
    w = w - (1/m) * learning_rate * (np.dot(x.transpose(),(saida-y)))

    custo_historico[i]=cal_custo(w,x,y)

  return w,custo_historico
```

```python
w = np.random.randn(1,100)
```

```python
w, custo_historico = gradiente(x,y,w,learning_rate=0.1,iteration=1000)
```
